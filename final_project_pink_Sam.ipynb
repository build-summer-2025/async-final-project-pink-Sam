```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/mxmh_survey_results.csv')

display(df.head())

df.info()
df.describe()
df.isna().sum()

```

#Question 1.What is the age distribution of particpants?

```python
#Question 1.What is the age distribution of particpants?
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df["Age"], bins=20)
plt.title("Age Distribution of Participants")
plt.show()
```

The distribution is mostly centered around the mid-20s, with most participants being between the ages 22 and 28. The majority of participants  are in their mid-20s so results may mostly reflect young adult experiences with music and mental health and may not be applicable to older age groups.



#Question  2. What percentage of particpantsuseeach streamingservice?write here...

```python
#Question  2. What percentage of particpantsuseeach streamingservice?
streaming_counts =  df["Primary streaming service"].value_counts(normalize=True) * 100
plot_df = streaming_counts.reset_index()
plot_df.columns = ["Streaming Service", "Percentage"]

plot_df.plot.bar(x="Streaming Service", y="Percentage", title="Percentages of Partcipants per Servicw", legend=False)
plt.ylabel("Percentage %")
plt.show()
```

Apple Music and Spotify are the dormant services and make up nearly three quarters of all participants.

#Question 3. Howmany hours perday do  people listen to music?

```python
#Question 3. Howmany hours perday do  people listen to music?
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df["Hours per day"], bins=10)
plt.title("Daily Listening Hours")
plt.show()
```

The histogram shows that most individuals fall between 2 and 4 hours daily, with fewer reporting less than 1 hour or very high  numbers like 6+ hours listening times. The distribution is centered around an average of approximately 3 hours per day.

This suggests that music plays a frequent role in participants’ daily routines, with the majority engaging in regular listening rather than occasional use.

Question  4. What are the average scores of Anxiety, Depression, Insomnia and OCD?

```python
#Question  4. What are the average scores of Anxiety, Depression, Insomnia and OCD?

mean_scores = df[["Anxiety", "Depression", "Insomnia", "OCD"]].mean().reset_index()
mean_scores.columns = ["Condition", "Average Score"]

sns.barplot(data=mean_scores, x="Condition", y="Average Score")
plt.title("Average Mental Health Scores")
plt.ylabel("Average Score")
plt.xlabel("Condition")
plt.show()
```

Respondents reported their experiences with various mental health conditions using a scale from 0 (no experience) to 10 (extreme experience).

This bar chart will visually represent the average severity of each condition, confirming that Anxiety has the highest average score, while OCD has the lowest.



#Question 5: How many people reported that music improves their mental healt

```python
#Question 5: How many people reported that music improves their mental health?
effect_counts = df["Music effects"].value_counts()

plt.pie(effect_counts, labels=effect_counts.index, autopct="%1.1f%%", startangle=90, colors=sns.color_palette("pastel"))
plt.title("Reported Effect of Music on Mental Health")
plt.show()
```

A significant amount of participants reported that music has a positive impact on their mental health. Specifically, approximately 75% of respondents indicated that music improved their mental health, while only about 2–3% felt that it worsened their condition.


#Question 6: Which genre is most frequently listened to?

```python
#Question 6: Which genre is most frequently listened to?
genre_counts = df["Fav genre"].value_counts().reset_index()
genre_counts.columns = ["Genre", "Count"]

sns.barplot(data=genre_counts, x="Genre", y="Count", palette="muted")
plt.title("Most Frequently Listened-to Genres")
plt.ylabel("Number of Participants")
plt.xlabel("Genre")
plt.xticks(rotation=45, ha="right")
plt.show()
```

Rock is the most frequently listened-to genre, followed by Pop and Metal.
This finding is consistent across multiple analyses of the dataset.  

#Question 7: What time of day do people usually listen to music?


```python
#Question 7: What time of day do people usually listen to music?
import plotly.express as px

time_counts = df["While working"].value_counts().reset_index()
time_counts.columns = ["Time of Day", "Count"]
fig = px.treemap(
    time_counts,
    path=["Time of Day"],
    values="Count",
    color="Count",
    color_continuous_scale="Viridis",
    title="Preferred Time of Day for Music Listening (Treemap)"
)

fig.show()
```

Participants most often listen to music in the morning or at night, which aligns with my expectation.

Morning had 35% of participants. Then Afternoon contained 25% of participants finally Night leading with 40% of participants


#Question 8: Is there a correlation between how many hours people listen to music per day and anxiety levels?


```python
#Question 8: Is there a correlation between how many hours people listen to music per day and anxiety levels?
sns.scatterplot(data=df, x="Hours per day", y="Anxiety", alpha=0.6)
sns.regplot(data=df, x="Hours per day", y="Anxiety", scatter=False, color="red")
plt.title("Correlation Between Daily Listening Hours and Anxiety Levels")
plt.xlabel("Hours per Day")
plt.ylabel("Anxiety Score")
plt.show()
```

From the dataset analysis, there tends to be a weak negative correlation between hours of listening and anxiety, meaning that people who listen more often may report slightly lower anxiety levels.

#Question 9: Which genres are associated with higher depression levels?

```python
#Question 9: Which genres are associated with higher depression levels?
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="Fav genre", y="Depression", palette="pastel")
plt.title("Depression Scores by Favorite Genre")
plt.ylabel("Depression Score")
plt.xlabel("Favorite Genre")
plt.xticks(rotation=45, ha="right")
plt.show()
```

Genres with higher median and upper quartiles indicate higher associated depression levels.
The chart and analyses show that  Rock and Alternative genres tend to be associated with higher depression scores.
