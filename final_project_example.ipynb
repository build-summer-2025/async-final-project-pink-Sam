```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("mxmh_survey_results.csv")

display(df.head())

df.info()
df.describe()
df.isna().sum()

#Question 1.What is the age distribution of particpants?
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df["Age"], bins=20)
plt.title("Age Distribution of Participants")
plt.show()

#Question  2. What percentage of particpantsuseeach streamingservice?
streaming_counts =  df["Primary streaming service"].value_counts(normalize=True) * 100
plot_df = streaming_counts.reset_index()
plot_df.columns = ["Streaming Service", "Percentage"]

plot_df.plot.bar(x="Streaming Service", y="Percentage", title="Percentages of Partcipants per Servicw", legend=False)
plt.ylabel("Percentage %")
plt.show()

#Question 3. Howmany hours perday do  people listen to music?
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df["Hours per day"], bins=10)
plt.title("Daily Listening Hours")
plt.show()

#Question  4. What are the average scores of Anxiety, Depression, Insomnia and OCD?

mean_scores = df[["Anxiety", "Depression", "Insomnia", "OCD"]].mean().reset_index()
mean_scores.columns = ["Condition", "Average Score"]

sns.barplot(data=mean_scores, x="Condition", y="Average Score")
plt.title("Average Mental Health Scores")
plt.ylabel("Average Score")
plt.xlabel("Condition")
plt.show()

#Question 5: How many people reported that music improves their mental health?
effect_counts = df["Music effects"].value_counts()

plt.pie(effect_counts, labels=effect_counts.index, autopct="%1.1f%%", startangle=90, colors=sns.color_palette("pastel"))
plt.title("Reported Effect of Music on Mental Health")
plt.show()

#Question 6: Which genre is most frequently listened to?
genre_counts = df["Fav genre"].value_counts().reset_index()
genre_counts.columns = ["Genre", "Count"]

sns.barplot(data=genre_counts, x="Genre", y="Count", palette="muted")
plt.title("Most Frequently Listened-to Genres")
plt.ylabel("Number of Participants")
plt.xlabel("Genre")
plt.xticks(rotation=45, ha="right")
plt.show()
```
